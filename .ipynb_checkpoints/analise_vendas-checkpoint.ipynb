### Carregamento de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# Formatando a visualização dos valores float
pd.options.display.float_format = '{:20,.2f}'.format
```

### Carregando dados a serem analisados

```python
df = pd.read_excel('files/AdventureWorks.xlsx')
df.sample(6)
```

### Visualizando formato e tipo dos dados

```python
print(f'Shape: {df.shape}')
print(f'Data type:\n{df.dtypes}')
```

### Verificando existência de missing values

```python
df.isnull().sum()
```

## Avaliando as vendas

```python
receita = df['Valor Venda'].sum()
custo = df['Custo Unitário'].mul(df['Quantidade']).sum()
lucro = receita - custo

print(f'Receita total: R$ {round(receita, 2)}')
print(f'Custo total: R$ {round(custo, 2)}')
print(f'Lucro total: R$ {round(lucro, 2)}')
```

```python
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
df['Lucro'] = df['Valor Venda'] - df['Custo']

df.sample(1)
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro anual (2008-2009)', color='c')
plt.xlabel('Ano', fontsize=15)
plt.ylabel('Lucro (Mi R$)', fontsize=15);
```

```python
df_08 = df[df['Data Venda'].dt.year == 2008] # Selecionando vendas de 2008
df_09 = df[df['Data Venda'].dt.year == 2009] # Selecionando vendas de 2009

fig = plt.subplots(nrows=1, ncols=1, figsize=(12, 4), sharex=True)
df_08.groupby(df['Data Venda'].dt.month)['Lucro'].sum().plot(label='2008')
df_09.groupby(df['Data Venda'].dt.month)['Lucro'].sum().plot(label='2009')
plt.title('Lucro mensal', fontsize=20)
plt.xlabel('Mês', fontsize=15)
plt.ylabel('Lucro (R$)', fontsize=15)
plt.legend();
```

## Avaliando produtos e marcas

```python
fig = plt.subplots(nrows=1, ncols=1, figsize=(12, 4), sharex=True)
df.groupby(df['Produto'])['Quantidade'].sum().sort_values().plot.barh(color='c')
plt.title('Total de produtos vendidos', fontsize=20)
plt.xlabel('Quantidade', fontsize=15)
plt.ylabel('Produto', fontsize=15);
```

```python
fig = plt.subplots(nrows=1, ncols=1, figsize=(12, 4), sharex=True)
df_08.groupby(df['Produto'])['Quantidade'].sum().sort_values().plot.barh(color='m')
plt.title('Total de produtos vendidos - 2008', fontsize=20)
plt.xlabel('Quantidade', fontsize=15)
plt.ylabel('Produto', fontsize=15);
```

```python
fig = plt.subplots(nrows=1, ncols=1, figsize=(12, 4), sharex=True)
df_09.groupby(df['Produto'])['Quantidade'].sum().sort_values().plot.barh(color='m')
plt.title('Total de produtos vendidos - 2009', fontsize=20)
plt.xlabel('Quantidade', fontsize=15)
plt.ylabel('Produto', fontsize=15);
```

```python
df.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro total por marca', color='c')
plt.xlabel('Marcas')
plt.ylabel('Lucro total (Mi R$)')
plt.xticks(rotation='horizontal');
```

```python
df_08.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro total por marca - 2008', color='m')
plt.xlabel('Marcas')
plt.ylabel('Lucro total (Mi R$)')
plt.xticks(rotation='horizontal');
```

```python
df_09.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro total por marca - 2009', color='m')
plt.xlabel('Marcas')
plt.ylabel('Lucro total (Mi R$)')
plt.xticks(rotation='horizontal');
```
